Instrucciones:

* Carga el dataset "penguins" de Seaborn. Utiliza df = sns.load_dataset("penguins")
* Usa plt.subplots() para crear una figura con 2 filas y 2 columnas.
* Genera los siguientes gráficos:
  * (0,0): sns.histplot(): Histograma de las longitudes de aletas.
  * (0,1): sns.boxplot(): Boxplot comparando la masa corporal por especie.
  * (1,0): sns.scatterplot(): Dispersión entre masa corporal y largo de aletas.
  * (1,1): sns.violinplot(): Violin plot de la masa corporal según la especie.
* Personaliza los gráficos con títulos y colores adecuados.
* Ajusta el diseño con plt.tight_layout().


```python
#import section

import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

df = sns.load_dataset("penguins")
#show data created
print(df)

fig, axes = plt.subplots(2,2, figsize=(15, 5))
# Set Seaborn style for better aesthetics
sns.set_theme()

# Plot on the first subplot (axes[0])
sns.histplot(data=df, x='flipper_length_mm', kde=True, ax=axes[0,0], color='cornflowerblue')
axes[0,0].set_title('Longitudes de Aletas')
axes[0,0].set_xlabel('Largo de aleta en mm')
axes[0,0].set_ylabel('Cuenta')

# Plot on the first subplot (axes[0])
sns.boxplot(data=df, x='species', y='body_mass_g', ax=axes[0,1], color='orchid')
axes[0,1].set_title('Masa corporal por especie')
axes[0,1].set_xlabel('Especie')
axes[0,1].set_ylabel('Masa Corporal')

# Plot on the first subplot (axes[0])
#new_df = df[['flipper_length_mm','body_mass_g']]
#df_melt = new_df.melt( value_vars=['flipper_length_mm', 'body_mass_g'],
#                       var_name='Value_Type',
#                       value_name='Value')
#print(df_melt)
sns.scatterplot(data=df, x='flipper_length_mm', y='body_mass_g',  ax=axes[1,0],color='blueviolet')
axes[1,0].set_title('Masa Corporal vs Largo de aletas')
axes[1,0].set_xlabel('Largo de aleta en mm')
axes[1,0].set_ylabel('Masa Corporal')

# Plot on the first subplot (axes[0])
sns.violinplot(data=df, x='species', y='body_mass_g', ax=axes[1,1], color='turquoise')
axes[1,1].set_title('Masa corporal por especie')
axes[1,1].set_xlabel('Especie')
axes[1,1].set_ylabel('Masa Corporal')


# Adjust layout to prevent overlapping titles/labels
plt.tight_layout()

# Display the plot
plt.show()
```
