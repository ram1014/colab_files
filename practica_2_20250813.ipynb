##Code to pull and process data from informe_acciones.txt

````python
#import section
import re
#Open File mbox.txt as read-only
f = open('/content/drive/MyDrive/ClassFiles/informe_acciones.txt', 'r', encoding='latin-1')

#loop all lines
for line in f:
  line = line.rstrip()
  #if "Asunto" get the achronym of the Company
  if re.search('^Asunto:',line):
    res = re.findall('^Asunto: Informe de Acción - (.*)$',line)
    if res:
      comp = res[0]
  #if "Fecha" get the date
  if re.search('^Fecha:',line):
    res = re.findall(r'^Fecha: (\d{4}-\d{2}-\d{2} \d{2}:\d{2}:\d{2})*$',line)
    if res:
      date = res[0]
  #if  "Valor actual" get the share value
  #in the sequence share value is the last one, print all values.
  if re.search('^Valor actual:',line):
    res = re.findall(r'^Valor actual: \$([0-9]+\.[0-9]+)*$',line)
    if res:
      val = res[0]
    print(f"En {date} el valor de Acción para {comp} es ${val}")
````
