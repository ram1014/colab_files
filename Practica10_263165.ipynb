Instrucciones:

1. Genera un DataFrame de ventas con Mes, Ventas, y Gastos. Utiliza randint para generar números aleatorios enteros. Para ventas los valores deben estar entre 2000 y 5000. Para gastos, los valores debn estar entre 1000 y 4000.
2. Usa GridSpec para organizar los subplots de la siguiente manera:
     * Primer gráfico (barra, ocupando dos columnas)
     * Segundo gráfico (línea, una columna)
     * Tercer gráfico (dispersión, dos filas).
3. Agrega títulos y ajusta espacios.


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec

#Create dataframe with the data to use
r_data = { 'Mes': ['Ene', 'Feb', 'Mar', 'Abr', 'May', 'Jun', 'Jul',
                   'Ago', 'Sep', 'Oct', 'Nov', 'Dic'],
           'Ventas': np.random.randint(2000,5000, size=12),
           'Gastos': np.random.randint(1000,4000, size=12)}
df = pd.DataFrame(r_data)

#show data created
print(df)


fig = plt.figure(figsize=(12, 10))  # Crear figura
gs = gridspec.GridSpec(4, 2)  # Definir una cuadrícula de 2x4

# Crear subplots en posiciones específicas
ax1 = fig.add_subplot(gs[0, :])  # Primer subplot (fila completa)
ax2 = fig.add_subplot(gs[1, 0])  # Segundo subplot (fila 1, columna 0)
ax3 = fig.add_subplot(gs[2:, :])  # Tercer subplot ocupa toda las ultimas 2 filas

#Bar plot
ax1.bar(df['Mes'], df['Ventas'], label='Ventas', color="slateblue")
ax1.bar(df['Mes'], df['Gastos'], label='Gastos', color="orangered")
ax1.set_title("Ventas vs Gastos Barras")
ax1.legend()
#Lines plot
ax2.plot(df['Mes'], df['Ventas'], label='Ventas', color="slateblue")
ax2.plot(df['Mes'], df['Gastos'], label='Gastos', color="orangered")
ax2.set_title("Ventas vs Gastos Lineas")
ax2.legend()
#Scatter plot
ax3.scatter(df['Mes'], df['Ventas'], label='Ventas', color="slateblue")
ax3.scatter(df['Mes'], df['Gastos'], label='Gastos', color="orangered")
ax3.set_title("Ventas vs Gastos Dispersion")
ax3.legend()
plt.tight_layout()
plt.show()

```
