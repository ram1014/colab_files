#Practica 9

Utilizar plt.subplots() para visualizar diferentes distribuciones generadas con NumPy y almacenadas en un DataFrame.

Instrucciones:

* Genera un DataFrame con Pandas que contenga 1000 valores en cuatro columnas (Normal, Uniforme, Exponencial, Poisson).
* Crea una figura con plt.subplots(2, 2) para organizar los histogramas en una cuadrícula de 2x2.
* Grafica cada distribución en un subplot usando ax[i, j].hist().
* Agrega títulos a cada subplot y ajusta la disposición con tight_layout().


```python
#Import Section
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#Create DataFrame
r_data = { 'Normal': np.random.randn(1000),
           'Uniforme': np.random.uniform(-2,2,1000),
           'Exponencial': np.random.exponential(1,1000),
           'Poisson': np.random.poisson(3,1000)}
df = pd.DataFrame(r_data)
#Print dataframe
print(df)

#create subplots object
fig, ax = plt.subplots(nrows=2, ncols=2)


#specify each plot using indexes
#Adding Normal
ax[0, 0].hist(df['Normal'], bins=30, color="cornflowerblue", alpha=0.7)
ax[0, 0].set_title("Distribución Normal")
#Adding Uniform
ax[0, 1].hist(df['Uniforme'], bins=30, color="turquoise", alpha=0.7)
ax[0, 1].set_title("Distribución Uniforme")
#Adding Exponencial
ax[1, 0].hist(df['Exponencial'], bins=30, color="hotpink", alpha=0.7)
ax[1, 0].set_title("Distribución Exponencial")
#Adding Poisson
ax[1, 1].hist(df['Poisson'], bins=30, color="wheat", alpha=0.7)
ax[1, 1].set_title("Distribución Poisson")

plt.tight_layout()
plt.show()
```
